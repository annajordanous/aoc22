# I--- Day 12: Hill Climbing Algorithm ---
You try contacting the Elves using your handheld device, but the river you're following must be too low to get a decent signal.

You ask the device for a heightmap of the surrounding area (your puzzle input). The heightmap shows the local area from above broken into a grid; the elevation of each square of the grid is given by a single lowercase letter, where a is the lowest elevation, b is the next-lowest, and so on up to the highest elevation, z.

Also included on the heightmap are marks for your current position (S) and the location that should get the best signal (E). Your current position (S) has elevation a, and the location that should get the best signal (E) has elevation z.


```python
def readin(filename):
    with open(filename) as f:
        lines = f.readlines()

    #something to remove \n 
    str_lines = []
    for l in lines:
        if l[-1:]=="\n":
            str_lines.append(l[0:-1])
        else: 
            str_lines.append(l)

    #check all was fine
    print(lines[-2:])
    print(str_lines[-2:])
    return str_lines

elf_data = readin("Day12input.txt")
```

```python
testd = ["Sabqponm",
"abcryxxl",
"accszExk",
"acctuvwj",
"abdefghi"]
```

```python

def findstart(dat, tokens):
    max_r = len(dat)
    max_c = len(dat[0])
    r = 0
    c = 0
    starts = []
    for r in range(0, max_r):
        for c in range(0, max_c):
            if dat[r][c] in tokens:
                starts.append((r, c))
    return starts

start = findstart(testd, ["S"])
print(start)
```

```python
def edge(a, b):
    # Your current position (S) has elevation a, and 
    # the location that should get the best signal (E) has elevation z.
    if a=="S":
        a = "a"
    if b=="E":
        b  = "z"
    return ord(b) <= ord(a)+1 

def efound(a):
    return a == "E"


def bfs(data, start): 
    
    dirs = {}
    dirs["up"] = (-1,0)
    dirs["down"] = (1,0)
    dirs["left"] = (0,-1)
    dirs["right"] = (0,1)


    queue = [(start, 0)]

    visited = set() 
    visited.add(start)

    found = False

    while not found and len(queue) > 0:
        (cur, counter) = queue.pop(0)  
        for d in dirs:
            neighbour_d = ((cur[0]+dirs[d][0]), (cur[1]+dirs[d][1]))
                # (sensible co-ordinate?)
            if neighbour_d[0] in range(0, len(data)) and neighbour_d[1] in range(0, len(data[0])):
                if edge(data[cur[0]][cur[1]], data[neighbour_d[0]][neighbour_d[1]]): # it's a neighbour
                        # print("check", a, ord(a),  b, ord(b)
                    if neighbour_d not in visited: # it's unvisited
                        if efound(data[neighbour_d[0]][neighbour_d[1]]): # is it what we're looking for?
                            found = True
                        else:
                            queue.append((neighbour_d, counter+1))
                            visited.add(neighbour_d)
    if found:
        return counter+1
    else: 
        return None

start_point = findstart(testd, ["S"])
print("BFS ", bfs(testd, start_point[0]))
```

# --- Part Two ---

As you walk up the hill, you suspect that the Elves will want to turn this into a hiking trail. The beginning isn't very scenic, though; perhaps you can find a better starting point.

To maximize exercise while hiking, the trail should start as low as possible: elevation a. The goal is still the square marked E. However, the trail should still be direct, taking the fewest steps to reach its goal. So, you'll need to find the shortest path from any square at elevation a to the square marked E.

Again consider the example from above:

    Sabqponm
    abcryxxl
    accszExk
    acctuvwj
    abdefghi
    
Now, there are six choices for starting position (five marked a, plus the square marked S that counts as being at elevation a). If you start at the bottom-left square, you can reach the goal most quickly:

    ...v<<<<
    ...vv<<^
    ...v>E^^
    .>v>>>^^
    >^>>>>>^
This path reaches the goal in only 29 steps, the fewest possible.

What is the fewest steps required to move starting from any square with elevation a to the location that should get the best signal?


```python
start_point2 = findstart(elf_data, ["S", "a"])
print(len(start_point2))

options = set()
for s in start_point2:
    x = bfs(elf_data, s)
    if x is not None:
        options.add(x)
print(min(options))
```
