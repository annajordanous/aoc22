# I--- Day 12: Hill Climbing Algorithm ---
You try contacting the Elves using your handheld device, but the river you're following must be too low to get a decent signal.

You ask the device for a heightmap of the surrounding area (your puzzle input). The heightmap shows the local area from above broken into a grid; the elevation of each square of the grid is given by a single lowercase letter, where a is the lowest elevation, b is the next-lowest, and so on up to the highest elevation, z.

Also included on the heightmap are marks for your current position (S) and the location that should get the best signal (E). Your current position (S) has elevation a, and the location that should get the best signal (E) has elevation z.


```python
def readin(filename):
    with open(filename) as f:
        lines = f.readlines()

    #something to remove \n 
    str_lines = []
    for l in lines:
        if l[-1:]=="\n":
            str_lines.append(l[0:-1])
        else: 
            str_lines.append(l)

    #check all was fine
    #print(lines[-2:])
    #print(str_lines[-2:])
    return str_lines

elf_data = readin("Day12input.txt")
```

```python
testd = ["Sabqponm",
"abcryxxl",
"accszExk",
"acctuvwj",
"abdefghi"]
```

```python

def findstart(dat):
    found = False
    d = len(dat)
    r = 0
    c = 0
    while not found:
        if dat[r][c] == "S":
            found = True 
            return (r, c)
        elif c+1 < d:
            c = c+1
        elif r+1 < d:
            r = r+1
            c = 0
        else: 
            return None

start = findstart(testd)
print(start)
```

```python
# def checkedge(data, r, c, dir):
#     cur = data[r][c]
#     rb = r+dir[0]
#     cb = c+dir[1]
#     if rb in range(0, len(data) and c in range[0, data]):
#         return edge(data[r][c], data[rb][cb])
#     else: return False

def edge(a, b):
    # Your current position (S) has elevation a, and 
    # the location that should get the best signal (E) has elevation z.
    if a=="S":
        a = "a"
    if b=="E":
        b  = "z"
    return ord(b) <= ord(a)+1 

def efound(a):
    return a == "E"


def bfs(data): 
    
    dirs = {}
    dirs["up"] = (-1,0)
    dirs["down"] = (1,0)
    dirs["left"] = (0,-1)
    dirs["right"] = (0,1)

# looked up pseudocode for bfs on https://www.geeksforgeeks.org/breadth-first-search-or-bfs-for-a-graph/

#Declare a queue and insert the starting vertex.

    cur = findstart(data)
    queue = [(cur, 0)]

#Initialize a visited array and mark the starting vertex as visited.

    visited = set() 
    visited.add(cur)

    found = False
#Follow the below process till the queue becomes empty:

    while not found and len(queue) > 0:
        #Remove the first vertex of the queue.
        (cur, counter) = queue.pop(0)  
        #print(cur, data[cur[0]][cur[1]])
        #Mark that vertex as visited.
        visited.add(cur)
        #Insert all the unvisited neighbours of the vertex into the queue.
        for d in dirs:
            neighbour_d = ((cur[0]+dirs[d][0]), (cur[1]+dirs[d][1]))
                # (sensible co-ordinate?)
            if neighbour_d[0] in range(0, len(data)) and neighbour_d[1] in range(0, len(data[0])):
                if edge(data[cur[0]][cur[1]], data[neighbour_d[0]][neighbour_d[1]]): # it's a neighbour
                        # print("check", a, ord(a),  b, ord(b)
                    if neighbour_d not in visited: # it's unvisited
                        if efound(data[neighbour_d[0]][neighbour_d[1]]): # is it what we're looking for?
                            found = True
                        else:
                            queue.append((neighbour_d, counter+1))
    if found:
        return counter+1
    else: 
        return None

print("BFS ", bfs(testd))
print("BFS2 ", bfs(elf_data))
```
